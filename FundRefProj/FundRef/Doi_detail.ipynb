```python
ins = []
for i,j in zip(data.other_names, data.value):
    ins.append(i+[j])
```

```python
ins
```

```python
data["institute"] = ins
```

```python
data.head()
```

```python
doi_ins = dict(zip(data.id,data.institute))
```

```python
with open('doi_ins.json', 'w') as fp:
    json.dump(doi_ins, fp)
```

```python
with open('doi_ins.json') as f:
    doi_ii = json.load(f)
```
